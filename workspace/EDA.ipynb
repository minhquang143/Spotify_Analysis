```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
artists_df = pd.read_csv('../data/artists.csv')
albums_df = pd.read_csv('../data/albums.csv')
tracks_df = pd.read_csv('../data/tracks.csv')
audio_features_df = pd.read_csv('../data/audio_features.csv')
```

```python
def overview(datasets_name, df):
    print(f"--- {datasets_name.upper()} DATAFRAME ---")
    print(df.info())
    print(df.describe(include='all'))
    print(df.head(), "\n")
```

```python
overview('artists', artists_df)
```

```python
overview('albums', albums_df)
```

```python
overview('tracks', tracks_df)
```

```python
overview('audio_features', audio_features_df)
```

```python
datasets = {'artists': artists_df, 'albums': albums_df, 'tracks': tracks_df, 'audio_features': audio_features_df}
for name, df in datasets.items():
    print(f"Missing values in {name.upper()}:")
    print(df.isnull().sum(), "\n")
```

```python
# Visualization: Artist popularity vs followers
sns.scatterplot(data=artists_df, x='followers', y='popularity')
plt.title('Artist Popularity vs Followers')
plt.xlabel('Followers')
plt.ylabel('Popularity')
plt.show()
```

Biểu đồ tương quan giữa độ nổi tiếng và số lượng người theo dõi (Followers):
Biểu đồ scatter plot này cho thấy mối quan hệ giữa độ nổi tiếng (popularity) và followers của các nghệ sĩ.

Quan sát chính:

Đa số các nghệ sĩ có lượng followers từ 0 đến 50 triệu và có độ nổi tiếng (popularity) dao động từ 70 đến 100.
Một số nghệ sĩ có lượng followers rất cao (trên 100 triệu) cũng đạt độ nổi tiếng tối đa (~100).
Phần dữ liệu phía dưới cho thấy các nghệ sĩ với lượng followers thấp (< 10 triệu) nhưng có độ nổi tiếng trung bình (~50-70).

Phân tích:

Có vẻ như followers không phải là yếu tố duy nhất quyết định popularity. Một số nghệ sĩ có popularity cao nhưng lượng followers không lớn, có thể do họ mới nổi hoặc có bài hát bùng nổ gần đây.

Tuy nhiên, followers cao thường đi kèm với popularity cao, cho thấy sự liên kết giữa sự theo dõi và mức độ ảnh hưởng của nghệ sĩ.

```python
audio_features_df.hist(bins=20, figsize=(12, 10))
plt.suptitle('Distribution of Audio Features')
plt.show()
```

Biểu đồ phân phối các đặc trưng âm thanh (audio features):

Các quan sát quan trọng:

Danceability (khả năng nhảy múa) và energy (năng lượng) có dạng phân phối đối xứng, phần lớn các bài hát tập trung ở giá trị trung bình (~0.5-0.7).

Key (tông nhạc) và time_signature (nhịp điệu) có phân phối dạng rời rạc, do đây là các đặc trưng phân loại hơn là liên tục.

Loudness (độ lớn âm thanh) tập trung cao ở khoảng từ -20 đến -10 dB, cho thấy đa số bài hát không quá lớn.

Valence (cảm xúc tích cực) và speechiness (mức độ giọng nói) phân bố lệch, cho thấy đa số bài hát nghiêng về cảm xúc trung tính và ít có sự hiện diện của giọng nói rõ ràng.

Instrumentalness (tính không lời) và acousticness (âm thanh tự nhiên) có sự phân bố mạnh về phía giá trị thấp, cho thấy đa số bài hát chứa yếu tố điện tử hoặc giọng hát rõ rệt.

Ý nghĩa:

Những phân phối này phản ánh xu hướng âm nhạc hiện đại: năng lượng cao, dễ nhảy múa, và thiên về âm thanh tổng hợp hơn là tự nhiên.

Dữ liệu cho thấy hầu hết các bài hát trong tập dữ liệu là nhạc có lời với năng lượng và nhịp điệu trung bình.

```python
numerical_cols = audio_features_df.select_dtypes(include=['float64', 'int64']).columns
numerical_df = audio_features_df[numerical_cols]

# Calculate correlation matrix
corr = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap for Audio Features')
plt.show()
```

Heatmap tương quan giữa các đặc trưng âm thanh (audio features):

Quan sát chính:

Danceability có tương quan trung bình với valence (0.43), nghĩa là những bài hát dễ nhảy múa thường mang cảm xúc tích cực.

Energy có tương quan mạnh nhất với loudness (0.71), cho thấy các bài hát có năng lượng cao thường có âm lượng lớn.

Acousticness có tương quan âm mạnh với energy (-0.63) và loudness (-0.50), phản ánh rằng các bài hát với âm thanh tự nhiên thường ít năng lượng và âm lượng thấp hơn.

Instrumentalness có tương quan thấp với hầu hết các đặc trưng, ngoại trừ acousticness, cho thấy các bài hát không lời thường có yếu tố âm thanh tự nhiên hơn.

Phân tích:

Mối tương quan giữa danceability và valence phù hợp với thực tế rằng những bài hát sôi động dễ nhảy múa thường mang lại cảm xúc tích cực.

Acousticness và instrumentalness thấp chứng tỏ âm nhạc hiện đại chủ yếu tập trung vào các yếu tố kỹ thuật số và giọng hát.


```python
# Ensure consistent date formats
albums_df['release_date'] = pd.to_datetime(
    albums_df['release_date'], 
    format='%Y-%m-%d', 
    errors='coerce'  # Invalid formats become NaT
)

# Handle cases with only the year
albums_df['release_date'] = albums_df['release_date'].fillna(
    pd.to_datetime(albums_df['release_date'], format='%Y', errors='coerce')
)

# Extract the release year
albums_df['release_year'] = albums_df['release_date'].dt.year

# Plot the number of albums per year
albums_per_year = albums_df.groupby('release_year').size()
albums_per_year.plot(kind='bar', figsize=(12, 6), title='Albums Released per Year')
plt.xlabel('Year')
plt.ylabel('Number of Albums')
plt.show()
```

Biểu đồ số lượng album phát hành theo năm (Albums Released per Year):
Tổng quan:

Biểu đồ thể hiện số lượng album phát hành mỗi năm, với trục hoành là các năm từ 1964 đến 2024 và trục tung là số lượng album.
Có sự gia tăng rõ rệt về số lượng album phát hành từ khoảng năm 2000 trở đi.
Nhận xét:

Trước năm 2000, số lượng album phát hành mỗi năm tương đối thấp, có thể phản ánh thời kỳ đầu của ngành công nghiệp âm nhạc hiện đại và giới hạn trong công nghệ sản xuất, phân phối.
Từ năm 2000 trở đi, số lượng album tăng mạnh, đặc biệt từ 2010 trở đi, cho thấy sự bùng nổ của nền công nghiệp âm nhạc, sự phổ biến của các nền tảng streaming và sự gia tăng của các nghệ sĩ độc lập.
Năm gần đây (2024), số lượng album đạt đỉnh, có thể do các yếu tố như: việc phát hành album trở nên dễ dàng hơn với công nghệ, sự đa dạng hóa trong sáng tạo nghệ thuật.

```python
# Kết hợp tracks và artists
merged_df = tracks_df.merge(artists_df, left_on='artist_id', right_on='id')
# Tính số lượng bài hát và độ nổi tiếng trung bình cho từng nghệ sĩ
artist_stats = merged_df.groupby('name_y').agg(
    num_songs=('id_x', 'count'),
    avg_popularity=('popularity_x', 'mean')
).reset_index()
# Tính tương quan
correlation = artist_stats['num_songs'].corr(artist_stats['avg_popularity'])
print(f"Hệ số tương quan giữa số lượng bài hát và độ nổi tiếng: {correlation:.2f}")

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
sns.scatterplot(data=artist_stats, x='num_songs', y='avg_popularity', alpha=0.7, color='blue')
plt.title('Mối Quan Hệ Giữa Số Lượng Bài Hát và Độ Nổi Tiếng')
plt.xlabel('Số Lượng Bài Hát')
plt.ylabel('Độ Nổi Tiếng Trung Bình')
plt.grid()
plt.show()
```

Biểu đồ mối quan hệ giữa số lượng bài hát và độ nổi tiếng trung bình (Mối Quan Hệ Giữa Số Lượng Bài Hát và Độ Nổi Tiếng):
Tổng quan:

Biểu đồ dạng scatter plot, trục hoành là số lượng bài hát và trục tung là độ nổi tiếng trung bình.
Các điểm phân bố khá đều, không có mối quan hệ tuyến tính rõ ràng giữa số lượng bài hát và độ nổi tiếng trung bình.
Nhận xét:

Có sự tập trung ở mức độ nổi tiếng trung bình từ 30 đến 50 và số lượng bài hát từ 50 đến 200, cho thấy phần lớn nghệ sĩ có mức độ nổi tiếng vừa phải và số lượng bài hát không quá cao.
Một số điểm rời rạc (outlier) ở phía trên bên phải cho thấy một số nghệ sĩ có số lượng bài hát rất lớn (300–500 bài) nhưng độ nổi tiếng vẫn chỉ ở mức trung bình.
Điều này cho thấy rằng số lượng bài hát không phải là yếu tố quyết định chính đến độ nổi tiếng. Yếu tố như chất lượng, quảng bá, và sự phù hợp với thị hiếu khán giả có thể ảnh hưởng lớn hơn.
