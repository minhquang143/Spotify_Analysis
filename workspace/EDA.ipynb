```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
artists_df = pd.read_csv('../data/artists.csv')
albums_df = pd.read_csv('../data/albums.csv')
tracks_df = pd.read_csv('../data/tracks.csv')
audio_features_df = pd.read_csv('../data/audio_features.csv')
```

Pháº§n dá»¯ liá»‡u phÃ­a dÆ°á»›i cho tháº¥y cÃ¡c nghá»‡ sÄ© vá»›i lÆ°á»£ng followers tháº¥p (< 10 triá»‡u) nhÆ°ng cÃ³ Ä‘á»™ ná»•i tiáº¿ng trung bÃ¬nh (~50-70).

Biá»ƒu Ä‘á»“ phÃ¢n phá»‘i cÃ¡c Ä‘áº·c trÆ°ng Ã¢m thanh (audio features):

CÃ¡c quan sÃ¡t quan trá»ng:

Danceability (kháº£ nÄƒng nháº£y mÃºa) vÃ  energy (nÄƒng lÆ°á»£ng) cÃ³ dáº¡ng phÃ¢n phá»‘i Ä‘á»‘i xá»©ng, pháº§n lá»›n cÃ¡c bÃ i hÃ¡t táº­p trung á»Ÿ giÃ¡ trá»‹ trung bÃ¬nh (~0.5-0.7).

Key (tÃ´ng nháº¡c) vÃ  time_signature (nhá»‹p Ä‘iá»‡u) cÃ³ phÃ¢n phá»‘i dáº¡ng rá»i ráº¡c, do Ä‘Ã¢y lÃ  cÃ¡c Ä‘áº·c trÆ°ng phÃ¢n loáº¡i hÆ¡n lÃ  liÃªn tá»¥c.

Loudness (Ä‘á»™ lá»›n Ã¢m thanh) táº­p trung cao á»Ÿ khoáº£ng tá»« -20 Ä‘áº¿n -10 dB, cho tháº¥y Ä‘a sá»‘ bÃ i hÃ¡t khÃ´ng quÃ¡ lá»›n.

Valence (cáº£m xÃºc tÃ­ch cá»±c) vÃ  speechiness (má»©c Ä‘á»™ giá»ng nÃ³i) phÃ¢n bá»‘ lá»‡ch, cho tháº¥y Ä‘a sá»‘ bÃ i hÃ¡t nghiÃªng vá» cáº£m xÃºc trung tÃ­nh vÃ  Ã­t cÃ³ sá»± hiá»‡n diá»‡n cá»§a giá»ng nÃ³i rÃµ rÃ ng.

Instrumentalness (tÃ­nh khÃ´ng lá»i) vÃ  acousticness (Ã¢m thanh tá»± nhiÃªn) cÃ³ sá»± phÃ¢n bá»‘ máº¡nh vá» phÃ­a giÃ¡ trá»‹ tháº¥p, cho tháº¥y Ä‘a sá»‘ bÃ i hÃ¡t chá»©a yáº¿u tá»‘ Ä‘iá»‡n tá»­ hoáº·c giá»ng hÃ¡t rÃµ rá»‡t.

Ã nghÄ©a:

Nhá»¯ng phÃ¢n phá»‘i nÃ y pháº£n Ã¡nh xu hÆ°á»›ng Ã¢m nháº¡c hiá»‡n Ä‘áº¡i: nÄƒng lÆ°á»£ng cao, dá»… nháº£y mÃºa, vÃ  thiÃªn vá» Ã¢m thanh tá»•ng há»£p hÆ¡n lÃ  tá»± nhiÃªn.

Dá»¯ liá»‡u cho tháº¥y háº§u háº¿t cÃ¡c bÃ i hÃ¡t trong táº­p dá»¯ liá»‡u lÃ  nháº¡c cÃ³ lá»i vá»›i nÄƒng lÆ°á»£ng vÃ  nhá»‹p Ä‘iá»‡u trung bÃ¬nh.

```python
numerical_cols = audio_features_df.select_dtypes(include=['float64', 'int64']).columns
numerical_df = audio_features_df[numerical_cols]

# Calculate correlation matrix
corr = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap for Audio Features')
plt.show()
```

Heatmap tÆ°Æ¡ng quan giá»¯a cÃ¡c Ä‘áº·c trÆ°ng Ã¢m thanh (audio features):

Quan sÃ¡t chÃ­nh:

Danceability cÃ³ tÆ°Æ¡ng quan trung bÃ¬nh vá»›i valence (0.43), nghÄ©a lÃ  nhá»¯ng bÃ i hÃ¡t dá»… nháº£y mÃºa thÆ°á»ng mang cáº£m xÃºc tÃ­ch cá»±c.

Energy cÃ³ tÆ°Æ¡ng quan máº¡nh nháº¥t vá»›i loudness (0.71), cho tháº¥y cÃ¡c bÃ i hÃ¡t cÃ³ nÄƒng lÆ°á»£ng cao thÆ°á»ng cÃ³ Ã¢m lÆ°á»£ng lá»›n.

Acousticness cÃ³ tÆ°Æ¡ng quan Ã¢m máº¡nh vá»›i energy (-0.63) vÃ  loudness (-0.50), pháº£n Ã¡nh ráº±ng cÃ¡c bÃ i hÃ¡t vá»›i Ã¢m thanh tá»± nhiÃªn thÆ°á»ng Ã­t nÄƒng lÆ°á»£ng vÃ  Ã¢m lÆ°á»£ng tháº¥p hÆ¡n.

Instrumentalness cÃ³ tÆ°Æ¡ng quan tháº¥p vá»›i háº§u háº¿t cÃ¡c Ä‘áº·c trÆ°ng, ngoáº¡i trá»« acousticness, cho tháº¥y cÃ¡c bÃ i hÃ¡t khÃ´ng lá»i thÆ°á»ng cÃ³ yáº¿u tá»‘ Ã¢m thanh tá»± nhiÃªn hÆ¡n.

PhÃ¢n tÃ­ch:

Má»‘i tÆ°Æ¡ng quan giá»¯a danceability vÃ  valence phÃ¹ há»£p vá»›i thá»±c táº¿ ráº±ng nhá»¯ng bÃ i hÃ¡t sÃ´i Ä‘á»™ng dá»… nháº£y mÃºa thÆ°á»ng mang láº¡i cáº£m xÃºc tÃ­ch cá»±c.

Acousticness vÃ  instrumentalness tháº¥p chá»©ng tá» Ã¢m nháº¡c hiá»‡n Ä‘áº¡i chá»§ yáº¿u táº­p trung vÃ o cÃ¡c yáº¿u tá»‘ ká»¹ thuáº­t sá»‘ vÃ  giá»ng hÃ¡t.


```python
# Ensure consistent date formats
albums_df['release_date'] = pd.to_datetime(
    albums_df['release_date'], 
    format='%Y-%m-%d', 
    errors='coerce'  # Invalid formats become NaT
)

# Handle cases with only the year
albums_df['release_date'] = albums_df['release_date'].fillna(
    pd.to_datetime(albums_df['release_date'], format='%Y', errors='coerce')
)

# Extract the release year
albums_df['release_year'] = albums_df['release_date'].dt.year

# Plot the number of albums per year
albums_per_year = albums_df.groupby('release_year').size()
albums_per_year.plot(kind='bar', figsize=(12, 6), title='Albums Released per Year')
plt.xlabel('Year')
plt.ylabel('Number of Albums')
plt.show()
```

Biá»ƒu Ä‘á»“ sá»‘ lÆ°á»£ng album phÃ¡t hÃ nh theo nÄƒm (Albums Released per Year):
Tá»•ng quan:

Biá»ƒu Ä‘á»“ thá»ƒ hiá»‡n sá»‘ lÆ°á»£ng album phÃ¡t hÃ nh má»—i nÄƒm, vá»›i trá»¥c hoÃ nh lÃ  cÃ¡c nÄƒm tá»« 1964 Ä‘áº¿n 2024 vÃ  trá»¥c tung lÃ  sá»‘ lÆ°á»£ng album.
CÃ³ sá»± gia tÄƒng rÃµ rá»‡t vá» sá»‘ lÆ°á»£ng album phÃ¡t hÃ nh tá»« khoáº£ng nÄƒm 2000 trá»Ÿ Ä‘i.
Nháº­n xÃ©t:

TrÆ°á»›c nÄƒm 2000, sá»‘ lÆ°á»£ng album phÃ¡t hÃ nh má»—i nÄƒm tÆ°Æ¡ng Ä‘á»‘i tháº¥p, cÃ³ thá»ƒ pháº£n Ã¡nh thá»i ká»³ Ä‘áº§u cá»§a ngÃ nh cÃ´ng nghiá»‡p Ã¢m nháº¡c hiá»‡n Ä‘áº¡i vÃ  giá»›i háº¡n trong cÃ´ng nghá»‡ sáº£n xuáº¥t, phÃ¢n phá»‘i.
Tá»« nÄƒm 2000 trá»Ÿ Ä‘i, sá»‘ lÆ°á»£ng album tÄƒng máº¡nh, Ä‘áº·c biá»‡t tá»« 2010 trá»Ÿ Ä‘i, cho tháº¥y sá»± bÃ¹ng ná»• cá»§a ná»n cÃ´ng nghiá»‡p Ã¢m nháº¡c, sá»± phá»• biáº¿n cá»§a cÃ¡c ná»n táº£ng streaming vÃ  sá»± gia tÄƒng cá»§a cÃ¡c nghá»‡ sÄ© Ä‘á»™c láº­p.
NÄƒm gáº§n Ä‘Ã¢y (2024), sá»‘ lÆ°á»£ng album Ä‘áº¡t Ä‘á»‰nh, cÃ³ thá»ƒ do cÃ¡c yáº¿u tá»‘ nhÆ°: viá»‡c phÃ¡t hÃ nh album trá»Ÿ nÃªn dá»… dÃ ng hÆ¡n vá»›i cÃ´ng nghá»‡, sá»± Ä‘a dáº¡ng hÃ³a trong sÃ¡ng táº¡o nghá»‡ thuáº­t.

```python
# Káº¿t há»£p tracks vÃ  artists
merged_df = tracks_df.merge(artists_df, left_on='artist_id', right_on='id')
# TÃ­nh sá»‘ lÆ°á»£ng bÃ i hÃ¡t vÃ  Ä‘á»™ ná»•i tiáº¿ng trung bÃ¬nh cho tá»«ng nghá»‡ sÄ©
artist_stats = merged_df.groupby('name_y').agg(
    num_songs=('id_x', 'count'),
    avg_popularity=('popularity_x', 'mean')
).reset_index()
# TÃ­nh tÆ°Æ¡ng quan
correlation = artist_stats['num_songs'].corr(artist_stats['avg_popularity'])
print(f"Há»‡ sá»‘ tÆ°Æ¡ng quan giá»¯a sá»‘ lÆ°á»£ng bÃ i hÃ¡t vÃ  Ä‘á»™ ná»•i tiáº¿ng: {correlation:.2f}")

# Váº½ biá»ƒu Ä‘á»“
plt.figure(figsize=(10, 6))
sns.scatterplot(data=artist_stats, x='num_songs', y='avg_popularity', alpha=0.7, color='blue')
plt.title('Má»‘i Quan Há»‡ Giá»¯a Sá»‘ LÆ°á»£ng BÃ i HÃ¡t vÃ  Äá»™ Ná»•i Tiáº¿ng')
plt.xlabel('Sá»‘ LÆ°á»£ng BÃ i HÃ¡t')
plt.ylabel('Äá»™ Ná»•i Tiáº¿ng Trung BÃ¬nh')
plt.grid()
plt.show()
```

Tá»•ng quan:

Biá»ƒu Ä‘á»“ dáº¡ng scatter plot, trá»¥c hoÃ nh lÃ  sá»‘ lÆ°á»£ng bÃ i hÃ¡t vÃ  trá»¥c tung lÃ  Ä‘á»™ ná»•i tiáº¿ng trung bÃ¬nh.
CÃ¡c Ä‘iá»ƒm phÃ¢n bá»‘ khÃ¡ Ä‘á»u, khÃ´ng cÃ³ má»‘i quan há»‡ tuyáº¿n tÃ­nh rÃµ rÃ ng giá»¯a sá»‘ lÆ°á»£ng bÃ i hÃ¡t vÃ  Ä‘á»™ ná»•i tiáº¿ng trung bÃ¬nh.
Nháº­n xÃ©t:
